## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns

y_df = attrition_df[['Attrition', 'Department']]
```

```python
# Create a list of at least 10 column names to use as X data
selected_columns = ['Age', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction', 'JobInvolvement',
                    'JobLevel', 'JobSatisfaction', 'YearsSinceLastPromotion', 'TrainingTimesLastYear', 'TotalWorkingYears']



# Create X_df using your selected columns
X_df = attrition_df[selected_columns]


# Show the data types for X_df
print(X_df.dtypes)

```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X_df, y_df, test_size=0.2, random_state=42)

#X = X_df.values
#y = y_df.values

#X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=42)

#print("X_train shape:", X_train.shape)
#print("X_test shape:", X_test.shape)
#print("y_train shape:", y_train.shape)
#print("y_test shape:", y_test.shape)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

```

```python
# Create a StandardScaler
scaler = StandardScaler()


# Fit the StandardScaler to the training data
scaler.fit(X_train)


# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop=None)
departments_train = y_train['Department'].values.reshape(-1, 1)
departments_test = y_test['Department'].values.reshape(-1, 1)

# Fit the encoder to the training data
encoder.fit(departments_train)


# Create two new variables by applying the encoder
# to the training and testing data
departments_train_encoded = encoder.transform(departments_train)
departments_test_encoded = encoder.transform(departments_test)

```

```python
# Create a OneHotEncoder for the Attrition column
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop=None)
attrition_train = y_train['Attrition'].values.reshape(-1, 1)
attrition_test = y_test['Attrition'].values.reshape(-1, 1)

# Fit the encoder to the training data
encoder.fit(attrition_train)


# Create two new variables by applying the encoder
# to the training and testing data
attrition_train_encoded = encoder.transform(attrition_train)
attrition_test_encoded = encoder.transform(attrition_test)
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
num_columns = X_train.shape[1]
print("Number of columns in X training data:", num_columns)

# Create the input layer
from tensorflow.keras.layers import Input, Dense
input_layer = Input(num_columns, name='input_layer')

# Create at least two shared layers
shared_layer1 = Dense(units=64, activation='relu')(input_layer)
shared_layer2 = Dense(units=128, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer


# Create the hidden layer
department_hidden_layer = Dense(units=32, activation='relu')(shared_layer2)

# Create the output layer
department_output_layer = Dense(departments_train_encoded.shape[1], activation='sigmoid', name='department_output')(department_hidden_layer)

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer


# Create the hidden layer
attrition_hidden_layer = Dense(units=32, activation='relu')(shared_layer2)


# Create the output layer
attrition_output_layer = Dense(attrition_train_encoded.shape[1], activation='sigmoid', name='attrition_output')(attrition_hidden_layer)

```

```python
# Create the model
from tensorflow.keras.optimizers import Adam
model = Model(inputs=input_layer, outputs=[department_output_layer, attrition_output_layer])

# Compile the model
model.compile(optimizer=Adam(), loss={'department_output': 'categorical_crossentropy', 'attrition_output': 'binary_crossentropy'}, metrics={'department_output':'accuracy', 'attrition_output':'accuracy'})


# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_scaled, {'department_output': departments_train_encoded, 'attrition_output': attrition_train_encoded}, epochs=100)


```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test, {'department_output': departments_test_encoded, 'attrition_output': attrition_test_encoded})
test_results
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {test_results[-2]}")
print(f"Attrition Accuracy: {test_results[-1]}")
```
